```python
os.chdir('/Users/lihaohan/Alpha_Research')
from decorators import do_on_dfs
@do_on_dfs
def boom_one(
    df: pd.DataFrame, 
    backsee: int = 5, 
    daily: bool = 0, 
    min_periods: int = None
    ) -> pd.DataFrame:

    if min_periods is None:
        min_periods = int(backsee * 0.5)
    if not daily:
        df_mean = (
            df.rolling(backsee, min_periods=min_periods).mean().resample("W").last()
        )
    else:
        df_mean = df.rolling(backsee, min_periods=min_periods).mean()
    return df_mean

@do_on_dfs
def var_one(
    df: pd.DataFrame, 
    backsee: int = 5, 
    daily: bool = 0, 
    min_periods: int = None
    ) -> pd.DataFrame:

    if min_periods is None:
        min_periods = int(backsee * 0.5)
    if not daily:
        df_var = (
            df.rolling(backsee, min_periods=min_periods).var().resample("W").last()
        )
    else:
        df_var = df.rolling(backsee, min_periods=min_periods).var()
    return df_var
```

```python
from tqdm import tqdm
mean_path = '/Volumes/T7Shield/Factor/factor/FactorZoo/mean'
var_path = '/Volumes/T7Shield/Factor/factor/FactorZoo/var'
for file in tqdm(files_rolling10):
    name = file.split('.')[0]
    try:
        df = pd.read_pickle(f'/Volumes/T7Shield/Factor/factor/FactorZoo/{file}')
        df = df.pivot(index='date', columns='code', values=df.columns[-1])
        factor_mean = boom_one(df, backsee=10, daily=1, min_periods=5)
        factor_var = var_one(df, backsee=10, daily=1, min_periods=5)
        factor_mean.to_pickle(os.path.join(mean_path, name + '10_mean.pickle'))
        factor_var.to_pickle(os.path.join(var_path, name + '10_var.pickle'))
    except Exception as e:
        print(f"Error processing {file}: {e}")
        continue
```

```python
@do_on_dfs
def standardlize(df: pd.DataFrame, all_pos: bool = 0) -> pd.DataFrame:
    """对因子dataframe做横截面z-score标准化

    Parameters
    ----------
    df : pd.DataFrame
        要做标准化的因子值, index是时间, columns是股票代码
    all_pos : bool, optional
        是否要将值都变成正数，通过减去截面的最小值实现, by default 0

    Returns
    -------
    pd.DataFrame
        标准化之后的因子
    """
    df = ((df.T - df.T.mean()) / df.T.std()).T
    if all_pos:
        df = (df.T - df.T.min()).T
    return df
```

```python
def cal_vov(vov1, vov2, window=40, min_periods=20):
    """
    计算风险模糊度
    Args:
    -------
        vov1: 日内5分钟收益率的波动率
        vov2: 日内5分钟收益率的Parkinson波动率
    """
    vov1 = vov1.pivot(index='date', columns='code', values=vov1.columns[-1])
    vov2 = vov2.pivot(index='date', columns='code', values=vov2.columns[-1])

    vov1, vov2 = standardlize(vov1), standardlize(vov2)

    if min_periods is None:
        min_periods = int(window * 0.5)

    vov_factor1 = vov1.rolling(window=window, min_periods=min_periods).std()
    vov_factor2 = vov2.rolling(window=window, min_periods=min_periods).std()

    return vov_factor1, vov_factor2
```

```python
vov1 = pd.read_pickle('/Volumes/T7Shield/Factor/factor/FactorZoo/vov1.pickle')
vov2 = pd.read_pickle('/Volumes/T7Shield/Factor/factor/FactorZoo/vov2.pickle')
vov_factor1, vov_factor2 = cal_vov(vov1, vov2, window=40, min_periods=20)
```
